```python
#asigno las variables categóricas de mi dataset

varCategoricas = ['Attrition_Flag','Gender','Education_Level','Marital_Status']

def pltCountplot(cats,data):    
    fig, axis = plt.subplots(len(cats) // 2,2, figsize=(24,11))  
    index = 0
    for i in range(len(cats) // 2):
        for j in range(2):            
            ax = sns.countplot(cats[index], data=data, ax=axis[i][j])            
            if cats[index] in ['Education_Level']:
                for item in ax.get_xticklabels():
                    item.set_rotation(15)
                
            for p in ax.patches:
                height = p.get_height()
                ax.text(p.get_x()+p.get_width()/2.,
                        height + 10,
                        '{:1.2f}%'.format(height/len(data)*100),
                        ha="center") 
            index += 1
```

#**Gráfico circular**

```python
#pie
leng = ['Python','R','Java','C++','FORTRAN']
porc = [35,25,20,15,5]
```

```python
fig = plt.figure(figsize=(6,6))
patches, texts, pcts= plt.pie(porc, 
        explode=(0, 0,0,0,0), 
        labels = leng, 
        autopct="%.1f%%", 
        wedgeprops = {"linewidth":3.0, "edgecolor":"white"},
        textprops = {"size": "x-large"}
        )
for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())

plt.setp(pcts, color="White", fontweight ="bold")
plt.title("Lenguajes de programación")
plt.show()
```

Todas las opciones para cmap se encuentra en: https://matplotlib.org/stable/tutorials/colors/colormaps.html

Otra alternatica es pcolormesh(), aqui debe definirse las coordenadas de una cuadrícula usando np.meshgrid(). Una función z que depende de x,y puede ser definida y graficada. 

```python
y, x = np.meshgrid(np.linspace(1, 5, 5), np.linspace(1, 5, 5))
x
```

```python
y, x = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
z = np.sin(x**2 + y**2) / (x**2 + y**2)
plt.pcolormesh(x, y, z)
plt.colorbar()
plt.show()
```

```python
plt.imshow(z)
```

contour() permite hacer curvas de nivel de la función anterior

```python
plt.contour(x,y,z)
```

# **Gráfico Cyberpunk**
Ya hemos visto que podemos personalizar los gráficos con Matplotlib. Veamos un ejemplo para crear un gráfico con estética cyberpunk. El siguiente notebook toma como referencia principal al artículo: Matplotlib Cyberpunk Style de Dominik Haitz https://matplotlib.org/matplotblog/posts/matplotlib-cyberpunk-style/ que se encuentra en el blog oficial de Matplotlib.

```python
import matplotlib.pyplot as plt
import pandas as pd

#Podemos ver todos los estilos disponibles
plt.style.available
```

```python
fig = plt.figure()
plt.style.use('dark_background')
dataDF = pd.DataFrame({"2021":[10,20,5,15,10], 
                       "2020":[8, 5, 6, 10, 6]})
dataDF.plot(marker = "o")
plt.show()
```

Para personalizar los gráficos es necesario saber cuales son los parámetros que podemos modificar, estos parámetros reciben el nombre de "rc parameters" https://matplotlib.org/stable/tutorials/introductory/customizing.html#customizing-with-matplotlibrc-files


```python
plt.rcParams
```

```python
plt.style.use('dark_background')
dataDF = pd.DataFrame({"2021":[10,20,5,15,10], 
                       "2020":[8, 5, 6, 10, 6]})
plt.rcParams["text.color"] = "0.8"
plt.rcParams["axes.labelcolor"] = "0.9"
plt.rcParams["figure.facecolor"] = "#212945"
colors = ['#08F7FE', '#FE53BB']
fig, ax = plt.subplots()
dataDF.plot(marker="o", color=colors, ax=ax)

plt.show()
```

```python
plt.rcParams['text.color'] = '0.9'
plt.rcParams['axes.labelcolor'] = '0.9'
plt.rcParams['xtick.color'] = '0.9'
plt.rcParams['ytick.color'] = '0.9'
plt.rcParams['figure.facecolor']= '#212946'
plt.rcParams['axes.facecolor']= '#212946'
plt.rcParams['savefig.facecolor']= '#212946'

colors = [
    '#08F7FE',  # cyan
    '#FE53BB',  # rosado
]

fig, ax = plt.subplots()
dataDF.plot(marker='o',color=colors,  ax=ax)
```

```python
plt.style.use("seaborn-dark")

for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
    plt.rcParams[param] = '#212946'  # bluish dark grey

for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = '0.9'  # very light grey

ax.grid(color='#2A3459') 
```
