#Introducción a Seaborn

Existen diferentes librerías para realizar gráficos en Python. Ya exploramos a Matplotlib, ahora es el turno de Seaborn. Esta librería está diseñada para realizar gráficos estadísticos y se puede integrar con Pandas. Su sitio web oficial es: https://seaborn.pydata.org/



En esta sesión exploraremos un dataset llamado "iris" que contiene longitudes de ancho y largo(sepal y petal) de la planta Iris (en 3 especies: Setos, Versicolor y Virginica). Por lo que este dataset es multivariable y requiere exploración. Puede ver los detalles en: https://archive.ics.uci.edu/ml/datasets/iris

```python
#importando seaborn
import seaborn as sns 
iris = sns.load_dataset("iris") #El dataset iris ya se encuentra incluido en la libreria
iris
```

El conjunto de datos "iris" es uno de tantos incluidos en la misma librería. Otros son penguins, tips, diamonds, geiser, etc.

```python
#Histogramaa que incluye a las 3 especies
sns.histplot(data=iris, x="sepal_width",hue='species',multiple="stack")
```

```python
sns.displot(iris, x="petal_length", hue="species",multiple="stack")
```

```python
#De este modo podemos tener los gráficos independientemente
sns.displot(data=iris, x="sepal_width", hue="species", col="species")
```

Los gráficos de KDE (kernel density estimation) hace una aproximación de la función de densidad de probabilidad que se ajusta a los datos. 

```python
sns.displot(iris, x="petal_length", kind="kde")
```

```python
sns.kdeplot(data=iris, x="sepal_width", hue="species", multiple="stack")
```

```python
sns.kdeplot(data=iris, x="petal_length", hue="species", multiple="stack")
```

```python
sns.displot(iris, x="petal_length", hue="species",kind="kde")
```

El siguiente muestra un gráfico de dispersión. Aqui podemos diferenciar por especie. Pruebe cambiando los atributos de x e y para observar si las especies se agrupan en cúmulos.

```python
sns.scatterplot(data=iris, x="petal_length", y="sepal_width", hue="species")
```

Joinplot combina dos gráficos. En la parte central observamos el gráfico de dispersión. Los ejes muestran la distribución univariada.

```python
#cambie los valores de x y
sns.jointplot(data=iris, x="petal_length", y="petal_width", hue="species")
```

Pairplot muestraa la relación de todas las variables involucradas en una forma de cuadrícula. 

```python
sns.pairplot(data=iris, hue="species")
```

Previamente, vimos como un gráfico de KDE describía la distribución de probabilidad para una variable. El siguiente gráfico muestra la distribución para dos variables.

```python
sns.displot(iris, x="sepal_length", y="sepal_width", kind="kde")
```

```python
#El gráfico por especies
sns.displot(iris, x="sepal_length", y="sepal_width", kind="kde",hue="species")
```

```python
#Combinando KDE y jointplot. Varie x e y
sns.jointplot(
    data=iris,
    x="sepal_length", y="sepal_width", hue="species",
    kind="kde"
)
```
